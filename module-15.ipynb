```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hours = [2 , 3 , 4, 1 , 5 , 7 ,3] 
days = [1 ,2 , 3 ,4 ,5 ,6 ,7] 

plt.plot(days,hours)
```

```python
plt.xlabel('Days')
plt.ylabel('Study Hours') 
plt.title('Trend of a student study time over a week')

plt.plot(days,hours)
```
